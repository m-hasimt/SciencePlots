```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_example_line_graph(fig, ax, n_vals=7):
    
    def model(x, p):
        return x ** (2 * p + 1) / (1 + x ** (2 * p))

    if n_vals <= 3:
        p_pattern = [10, 20, 50]
    elif n_vals <= 7:
        p_pattern = [5, 10, 15, 20, 30, 50, 100] # 7 pattern
    elif n_vals <= 9:
        p_pattern = [5, 7, 10, 15, 20, 30, 38, 50, 100]
    else:
        p_pattern = [5, 7, 10, 15, 20, 30, 38, 50, 100, 500]

    x = np.linspace(0.75, 1.25, 201)
    for p in p_pattern[0:n_vals]:
        ax.plot(x, model(x,p), label=p)
    ax.legend(title='Order')
    ax.set(xlabel='Voltage (mV)')
    ax.set(ylabel='Current ($\mu$A)')
    ax.autoscale(tight=True)
    fig.show()
    
```

## science style

```python
with plt.style.context(['science']):
    fig, ax = plt.subplots()
    plot_example_line_graph(fig, ax, n_vals=6)
```

### science + grid style

```python
with plt.style.context(['science', 'grid']):
    fig, ax = plt.subplots()
    plot_example_line_graph(fig, ax, n_vals=6)
```

### science + ieee style

```python
with plt.style.context(['science', 'ieee']):
    fig, ax = plt.subplots()
    plot_example_line_graph(fig, ax, n_vals=3)
```

### science + retro color

```python
with plt.style.context(['science', 'retro']):
    fig, ax = plt.subplots()
    plot_example_line_graph(fig, ax, n_vals=6)
```
